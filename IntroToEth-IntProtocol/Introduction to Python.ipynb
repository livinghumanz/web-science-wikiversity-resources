## Function

Functions take arguments, execute code and return results.

````python
def myfunc(x): # declaration and definition
    x = x**x
    return x
````

````python
myfunc(4)
````

````python
def divide(x,y): # two arguments
    return x/y
divide(50,5)
````

````python
def recursive(i): # recursive call
    if i < 0:
        return
    else:
        print(i)
        return recursive(i-1)
recursive(5)
````

````python
def even(*args): # taking list as input
    evn = []
    for i in args:
        if i%2==0:
            evn.append(i)
    return evn
even(1,2,3,4,50,23,123)
````

````python
def prettyPrint(**args): # taking dictionary as input
    print(args.items())
    return
prettyPrint(width = 2, height = 4)
````

### Map

Apply a function on values in list.

````python
items = [1, 2, 3, 4, 5] # input
squared = [] # future squared values
for i in items:
    squared.append(i**2) # execution
print(squared)
````

````python
squared = list(map(lambda x: x**2, items)) # map applies lambda on all list elements
print(squared)
````

# MatPlotLib

MatPlotLib is a library for plotting graphs. Examples are mostly taken from https://matplotlib.org/users/pyplot_tutorial.html

````python
import matplotlib.pyplot as plt
````

````python
plt.plot([1,2,3,4]) # define values
plt.ylabel('some numbers') # set title of y axis
plt.show()
````

````python
plt.plot([1,2,3,4], [1,4,9,16], 'o') # passing x and y coordinates
plt.axis([0, 6, 0, 20]) # axes limitation [xmin, xmax, ymin, ymax]
plt.show()
````

````python
t = np.arange(0, 5, 0.2) # evenly sampled time at 200ms intervals

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
````

````python
x = np.random.rand(500) # 500 values in [0, 1)
y = np.random.rand(500) # 500 values in [0, 1)
colors = (0,0,0) # custom coloring
size = 100 # size of points
 
# do scatter plot
plt.scatter(x, y, s=size, c=colors, alpha=0.3)
plt.title('scatter plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````

````python
labels = ["This", "That"]
data = [4, 7]

x = np.array(range(len(data))) # locations of bars on x axis
plt.bar(x, data, width=0.5) # plotting of bars
plt.xticks(x, labels) # labelng
plt.show()
````
