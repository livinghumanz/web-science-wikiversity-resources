## 1. Ethernet Frame (5 Points)



Find:
1. Source MAC Address **00 13 10 e8 dd 52**
2. Destination MAC Address **00 27 10 21 fa 48**
3. What protocol is inside the data payload? **08 06 (ARP)**
4. Please mention what the last 2 fields hold in the above frame. **__It Holds Target Address and Target Mac Address__**

#### Explanation: 
``` text
    Field              Hex Values                  Decimal / Meaning
Destination MAC     00 27 10 21 fa 48       The target hardware address
Source MAC          00 13 10 e8 dd 52       The sender's hardware address
EtherType           08 06                   ARP (Address Resolution Protocol)
```
- Because the Eth type is ARP the remaining data follows ARP structure

``` text

Field	            Hex Value	            Meaning
Hardware Type	    00 01                   Ethernet (1)
Protocol Type	    08 00                   IPv4 (0x0800)
Hardware Size	    06                      MAC addresses are 6 bytes
Protocol Size	    04                      IP addresses are 4 bytes
Opcode	            00 01                   Request (1)
Sender MAC          00 13 10 e8 dd 52       Sender's MAC (matches Source MAC)
Sender IP           c0 a8 02 01             192.168.2.1
Target MAC          00 00 00 00 00 00       Unknown (typical for a Request)
Target IP           c0 a8 02 67             192.168.2.103
```

This frame is a broadcast/unicast search. The device with the IP 192.168.2.1 is asking the network:

    "Who has the IP address 192.168.2.103? Tell me (the device at MAC 00:13:10:e8:dd:52) your MAC address."


## 3. Basic Network Tools (10 Points)
Overview of Network Commands
The following tools are essential for network diagnostics and discovery:

- `ipconfig / ifconfig:` Displays the network configuration of your local machine (IP address, Subnet Mask, Default Gateway).
- `ping:` Tests reachability to a host and measures round-trip time (latency).
- `traceroute:` Shows the path (hops) that a packet takes to reach a destination.
- `arp:` Displays the Address Resolution Protocol cache (maps IP addresses to MAC addresses).
- `dig:` A tool for interrogating DNS name servers (useful for checking IP records and query times).

#### Consider a situation in which you need to check if www.wikipedia.org is reachable or not. Using the knowledge you gained above to find the following information:
1. The % packet loss if at all it happened after sending 100 packets.
    > ping -c 100 www.wikipedia.org  --> 0% (5 packets transmitted, 5 received)
2. Size of the packet sent to Wikipedia server 
    > 56 bytes (ICMP payload) / 84 bytes (including headers)
3. IP address of your machine and the Wikipedia server
    > IP -Wiki : 185.15.59.224
4. Query Time for DNS query of the above url.
    > Query time : 101 msec (`$ dig www.wikipedia.org`)
5. Number of Hops in between your machine and the server
    > `$ traceroute www.wikipedia.org` 30hops
6. MAC address of the device that is acting as your network gateway.
    > `arp -a` Gateway-iP : 10.0.2.2  Gateway-MAC-Address: 52:55:0a:00:02:03  NetworkInterface: enp0s3










### 4. Simple Python Programming (10 Points)
Write a simple python program that does the following:
1. Generate a random number sequence of 10 values between 0 to 90.
2. Perform sine and cosine operation on numbers generated.
3. Store the values in two different arrays named SIN & COSIN respectively.
4. Plot the values of SIN & COSIN in two different colors.
5. The plot should have labeled axes and legend.

````python
# Generate a random No seq of 10 Values 0-99
import random
randomNumbers = [random.randint(0,99) for _ in range(0,10)]
print(randomNumbers)
````

````python
# Perform sine and cosine on the above generated random No seq
import math
SIN = [math.sin(num) for num in randomNumbers]
COSIN = [math.cos(num) for num in randomNumbers]
print("Sine Values: ", SIN)
print("Cosine Values: ", COSIN)
````

````python
# plot the values of sin and cosine in 2 different colors
import matplotlib.pyplot as plt
plt. plot(randomNumbers, SIN, 'r-', label='Sine')
plt. plot(randomNumbers, COSIN, 'b-', label='Cosine')
plt.legend()
plt.show()
````
