Write a simple python script that does the following:

```python
# read the dataset dataset.csv file as a pandas dataframe and print the first 5 rows
import pandas as pd
df = pd.read_csv('dataset.csv')

df.head()
```

```python
# plot the time distribution of the dataset using matplotlib
import matplotlib.pyplot as plt

plt.hist(df['time_in_seconds'], bins=20, color='blue', alpha=0.7)
plt.legend()
plt.show()
```
